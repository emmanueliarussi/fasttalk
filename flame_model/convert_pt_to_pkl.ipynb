```python
import torch
import pickle
import numpy as np

class Struct:
    def __init__(self, **entries):
        self.__dict__.update(entries)

# Cargar el .pt
pt_data = torch.load("assets/FLAME_with_eye.pt", map_location="cpu")

# Extraer el modelo
flame_dict = pt_data['flame_model']

# Convertir todos los tensores a NumPy arrays
flame_numpy = {k: v.numpy() if isinstance(v, torch.Tensor) else v for k, v in flame_dict.items()}

# Crear Struct con NumPy
flame_struct = Struct(**flame_numpy)

# Guardar como .pkl compatible
with open("assets/FLAME_with_eye.pkl", "wb") as f:
    pickle.dump(flame_struct.__dict__, f, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
import torch
import numpy as np

pt_data = torch.load("assets/FLAME_with_eye.pt", map_location="cpu")
flame = pt_data["flame_model"]

# Get vertices and faces
vertices = flame["v_template"].numpy()  # shape: (N, 3)
faces = flame["f"].numpy()              # shape: (F, 3)

# Write to OBJ file
with open("head_template_mesh_with_eye.obj", "w") as f:
    for v in vertices:
        f.write(f"v {v[0]} {v[1]} {v[2]}\n")
    for face in faces:
        # OBJ format uses 1-based indexing
        f.write(f"f {face[0]+1} {face[1]+1} {face[2]+1}\n")
```
